```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
from PIL import Image, ImageDraw
from matplotlib.animation import FuncAnimation
from io import BytesIO
import random
```

```python
def bubbleSort_basic(A, n, history):
  for i in range(n-1):
      for j in range(n-i-1):
          history.append({'arr': A.copy(), 'idx': j, 'target': j + 1})
          if (A[j] > A[j + 1]):
              A[j], A[j + 1] = A[j + 1], A[j]
              history.append({'arr': A.copy(), 'idx': j, 'target': j + 1})

def bubbleSort_optmized(A, n, history):
  for i in range(n-1):
    cnt = 0
    for j in range(n-i-1):
      history.append({'arr': A.copy(), 'idx': j, 'target': j + 1})
      if (A[j] > A[j + 1]):
        cnt += 1
        A[j], A[j + 1] = A[j + 1], A[j]
        history.append({'arr': A.copy(), 'idx': j, 'target': j + 1})
    if (cnt == 0):
      return

def selectionSort(A,n, history):
  for i in range(n-1,0,-1):
    idx = 0
    for j in range(i+1):
      history.append({'arr': A.copy(), 'idx': j, 'target': idx})
      if (A[idx] < A[j]):
        idx = j
    A[idx], A[i] = A[i], A[idx]
    history.append({'arr': A.copy(), 'idx': idx, 'target': i})

def insertionSort(A,n, history):
  for i in range(1,n):
    num = A[i]
    idx = i - 1
    history.append({'arr': A.copy(), 'idx': i, 'target': -1}) # Highlight element being picked to insert
    while (idx >= 0 and A[idx] > num):
      history.append({'arr': A.copy(), 'idx': idx, 'target': i}) # Highlight element being shifted
      A[idx + 1] = A[idx]
      idx -= 1
    A[idx + 1] = num
    history.append({'arr': A.copy(), 'idx': idx + 1, 'target': -1}) # Highlight element placed

def mergeSort(A, left, right, history):
    if (left >= right):
        return
    mid = (left + right) // 2;
    mergeSort(A, left, mid, history)
    mergeSort(A, mid + 1, right, history)
    merge(A, left, mid, right, history)

def merge(A,left,mid,right, history):
  array_left = []
  array_right = []
  i = 0
  j = 0
  for k in range(mid-left+1):
    array_left.append(A[left+k])
  for k in range(right-mid):
    array_right.append(A[k + mid + 1])

  history.append({'arr': A.copy(), 'idx': left, 'target': right}) # Initial state for merge

  while (i <= mid - left and j < right - mid):
    history.append({'arr': A.copy(), 'idx': left + i, 'target': mid + 1 + j}) # Highlight elements being compared
    if (array_left[i] <= array_right[j]):
      A[i + j + left] = array_left[i]
      i += 1
    else:
      A[i + j + left] = array_right[j]
      j += 1
    history.append({'arr': A.copy(), 'idx': i + j + left -1, 'target': -1}) # Highlight element just placed

  while (i <= mid - left):
    A[i + j + left] = array_left[i]
    i += 1
    history.append({'arr': A.copy(), 'idx': i + j + left -1, 'target': -1}) # Highlight element just placed

  while (j < right - mid):
    A[i + j + left] = array_right[j]
    j += 1
    history.append({'arr': A.copy(), 'idx': i + j + left -1, 'target': -1}) # Highlight element just placed

def quickSort(A, left, right, history):
  if (left < right):
    div_idx = partition(A, left, right, history)
    quickSort(A, left, div_idx - 1, history)
    quickSort(A, div_idx + 1, right, history)

def partition(A,left,right, history):
  div_idx = (left + right) // 2
  i = left
  j = right - 1

  history.append({'arr': A.copy(), 'idx': right, 'target': div_idx}) # Highlight pivot and its initial position
  A[div_idx], A[right] = A[right], A[div_idx] # Move pivot to end
  history.append({'arr': A.copy(), 'idx': right, 'target': div_idx}) # After swapping pivot to end

  pivot_val = A[right]

  while (i <= j):
    history.append({'arr': A.copy(), 'idx': i, 'target': right}) # Compare A[i] with pivot
    while (A[i] < pivot_val):
      i += 1
      history.append({'arr': A.copy(), 'idx': i, 'target': right}) # Move i

    history.append({'arr': A.copy(), 'idx': j, 'target': right}) # Compare A[j] with pivot
    while (A[j] >= pivot_val and j >= i):
      j -= 1
      history.append({'arr': A.copy(), 'idx': j, 'target': right}) # Move j

    if (i < j):
      A[i], A[j] = A[j], A[i]
      history.append({'arr': A.copy(), 'idx': i, 'target': j}) # After swap
      i += 1
      j -= 1
  A[i], A[right] = A[right], A[i] # Place pivot
  history.append({'arr': A.copy(), 'idx': i, 'target': right}) # After placing pivot
  return i

def heapSort(A,n, history):
  build_heap(A,n, history)
  for i in range(n-1,0,-1):
    history.append({'arr': A.copy(), 'idx': 0, 'target': i}) # Highlight root and last element before swap
    A[0],A[i] = A[i],A[0]
    history.append({'arr': A.copy(), 'idx': 0, 'target': i}) # Highlight after swap
    heapify(A,i,0, history)

def heapify(A, n, i, history):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Highlight current node and its children for comparison
    history.append({'arr': A.copy(), 'idx': i, 'target': left if left < n else -1})
    history.append({'arr': A.copy(), 'idx': i, 'target': right if right < n else -1})

    if (left < n and A[left] > A[largest]):
        largest = left
    if (right < n and A[right] > A[largest]):
        largest = right
    if (largest != i):
        history.append({'arr': A.copy(), 'idx': largest, 'target': i}) # Highlight elements before swap
        A[largest], A[i] = A[i], A[largest]
        history.append({'arr': A.copy(), 'idx': largest, 'target': i}) # Highlight elements after swap
        heapify(A, n, largest, history)

def build_heap(A,n, history):
  for i in range(n//2-1,-1,-1):
    heapify(A,n,i, history)

# 各フレームごとにグラフを更新
def update(frame_data):
    arr = frame_data['arr']
    idx = frame_data['idx']
    target = frame_data['target']
    for i, (bar, val) in enumerate(zip(bars, arr)):
        bar.set_height(val)
        if i == idx:
            bar.set_color('r') # 比較・交換される要素
        elif i == target:
            bar.set_color('y') # 比較・交換される要素
        else:
            bar.set_color('b') # その他の要素

def implement(A,optI,optJ):
  n = len(A)
  left = 0
  right = n-1
  history = [] # Initialize history here

  initial_arr = []
  match optI:
    case 0:
      initial_arr = [i + 1 for i in range(n)]
    case 1:
      initial_arr = [n - i for i in range(n)]
    case 2:
      initial_arr = [i + 1 for i in range(n)]
      random.shuffle(initial_arr)

  # 最初の配列の状態を追加する
  history.append({'arr': initial_arr.copy(), 'idx': -1, 'target': -1})

  A_copy = initial_arr.copy()

  match optJ:
    case 0:
      bubbleSort_basic(A_copy, n, history)
    case 1:
      bubbleSort_optmized(A_copy, n, history)
    case 2:
      selectionSort(A_copy, n, history)
    case 3:
      insertionSort(A_copy, n, history)
    case 4:
      mergeSort(A_copy, left, right, history)
    case 5:
      quickSort(A_copy, left, right, history)
    case 6:
      heapSort(A_copy, n, history)

  # ソート済みの状態を長めに表示したいので、グラフをループで追加する
  for i in range(50):
    history.append({'arr': A_copy.copy(), 'idx': -1, 'target': -1})

  return history
```

```python
n = 30
A = range(n)
data = []
str_algo = ["バブルソート(打ち切りなし)","バブルソート(打ち切りあり)","選択ソート","挿入ソート","マージソート","クイックソート","ヒープソート"]
str_opt = ["昇順","降順","ランダム"]
for i in range(3):
  for j in range(7):
    data = implement(A,i,j)
    fig, ax = plt.subplots()
    bars = ax.bar(range(n),data[0]['arr'],color = 'b',width = 1)
    ax.set_xlim(-0.5, n - 0.5)
    ax.set_ylim(0, max(A) * 1.1)
    ax.set_xticks([])
    ax.set_yticks([])
    show = FuncAnimation(fig,update,frames = data, blit=False,interval = 50)
    show.save(f"{str_algo[j]}({str_opt[i]}).gif",writer = "pillow",fps = 30)
    plt.close()
```
